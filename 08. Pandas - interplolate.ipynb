## Handling Missing Value using - Interpolate() Function
    Pnadas interpolate() function is basically used to fill NaN values in DataFrame and Series.
    Very Powerfull Function
    Uses various interpolation techniques

```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Dell\\Desktop\\_PANDAS_\\Data\\student_result1.csv")
df.head(2)
```

### Using interpolate function 

```python
df.interpolate()
```

### Using interpolate -  method = "Linear" 

```python
df.interpolate(method= 'linear')
```
