## Various Parameters of pd.read_csv() 

```python
import pandas as pd 
```

```python
df = pd.read_csv("C:\\Users\\Dell\\Desktop\\_PANDAS_\\Data\\Fortune_10_day13.csv")
df.head(3)
```

### Reads only a fixed number of rows from the file.

```python
df1 = pd.read_csv("C:\\Users\\Dell\\Desktop\\_PANDAS_\\Data\\Fortune_10_day13.csv",nrows=3)
df1
```

### Selects specific columns to load.

```python
df2 = pd.read_csv("C:\\Users\\Dell\\Desktop\\_PANDAS_\\Data\\Fortune_10_day13.csv",usecols=[0,1,2])
df2.head(3)
```

### Skips specific rows while reading the file.

```python
df3 = pd.read_csv("C:\\Users\\Dell\\Desktop\\_PANDAS_\\Data\\Fortune_10_day13.csv",skiprows=2) 
df3.head(3)
```

### Sets one column as the DataFrame index.

```python
df4 = pd.read_csv("C:\\Users\\Dell\\Desktop\\_PANDAS_\\Data\\Fortune_10_day13.csv",index_col=0) 
df4.head(3)
```

### Specifies which row contains column names.

```python
df5 = pd.read_csv("C:\\Users\\Dell\\Desktop\\_PANDAS_\\Data\\Fortune_10_day13.csv",header=4) 
df5.head(3)
```

### Provides custom column names.

```python
pd.read_csv('C:\\Users\\Dell\\Desktop\\_PANDAS_\\Data\\Fortune_10_day13.csv', names = ['1','2','3','4','5','6','7'] ).head(4)
```
