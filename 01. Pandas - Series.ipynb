## Basic of pandas Series
    A Pandas Series is a one-dimensional labeled array that can store data of any type 
    (integers, floats, strings, Python objects, etc.).
    It has two main components:
    Values → the actual data
    Index → labels for each value (like row numbers, but customizable)

```python
import pandas as pd
```

### version checking 

```python
pd.__version__
```

### Creating a pandas series

```python
pan_series = pd.Series([1,2,3,7,9,1,"Git",True])
pan_series
```

```python
type(pan_series)
```

### Creating a empty series

```python
pan_series1 = pd.Series([])
pan_series1
```

```python
# Changing the index of series
pan_series2 = pd.Series([1,2,3,7,9,1,"Git",True],index=['a','b','c','d','e','f','g','h'])
pan_series2
```

```python
pan_series3 = pd.Series([1,23,4,6,2,3],dtype=float)
pan_series3
```

### Creating series from scalar values scalar values means single value

```python
pan_series4 = pd.Series(3,index=[1,2,3,4,5,6,7,8])
pan_series4
```

###  Creating series from python dictionary

```python
pd_dic = {"Srno" : [1,2,3,4],"Roll_no":[101,102,103,104],"Subject": ["m1","m2","m3","m4"]}
pd_series5 = pd.Series(pd_dic)
pd_series5
```

```python
# Access the series element by index
pd_series5[2]
```

```python
# Slicing in series
pd_series5[1:]
```

### maths operation

```python
s5 = pd.Series([1,2,3,4,5])
s6 = pd.Series([1,2,3,4,5])
s5.add(s6)
```

```python
print("Mean of series s5 ::",s5.mean())
print("Max Element of series s5 ::",s5.max())
print("Samllest Element of series s5 ::",s5.min())
```
