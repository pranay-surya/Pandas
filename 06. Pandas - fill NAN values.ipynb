## Fill NaN Value
    Filling NaN (Missing) Values in Pandas
    In Pandas, NaN (Not a Number) represents missing or undefined data.
    The method fillna() is used to replace NaN values with a specified value or method instead of removing them.

    Most Common Way to Fill NAN Value 
     - ffill -> Fill NAN Values with Forward values
     - bfill -> Fill NAN values with Previous Values
     - etc

```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Dell\\Desktop\\_PANDAS_\\Data\\Fortune_10.csv")
df
```

### Fill the NaN value by dictionary

```python
df.fillna({'Industry':'None','Profit': 0,'Growth':0})  
```

### Fill NAN by method - ffill

```python
df.fillna(method = 'ffill',limit = 1,inplace=True)
df
```

### Fill NAN by method - bfill

```python
df.fillna(method = 'bfill',inplace = True)
df
```
